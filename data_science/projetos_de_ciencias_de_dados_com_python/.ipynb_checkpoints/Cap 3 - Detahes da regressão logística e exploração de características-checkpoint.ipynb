**Lembrete:**

* Não devemos usar a variável de gênero devido a questões éticas;
* Foi verificado que PAY_2, PAY_3, ..., PAY_6 estão incorretas e devem ser ignoradas;

Além disso, não examinaresmo a codificação one-hot que criamos a partir da variável EDUCATION, já que as informações dessas colunas já estão incluídas, pelo menos de certa forma, na característica original.

Para concluir, não faz sentido usar ID como característica. 

Assim, criaremos outra lista de nomes de colunas que não sejam características ou a resposta. Queremos excluir essas variáveis de nossa análise:

```python
items_to_remove = ['ID', 
                   'SEX', 
                   'PAY_2', 
                   'PAY_3', 
                   'PAY_4', 
                   'PAY_5', 
                   'PAY_6',
                   'EDUCATION_CAT', 
                   'graduate school', 
                   'high school',
                   'others',
                   'university']

items_to_remove
```

```python
features_response = [item for item in features_response if item not in items_to_remove]

features_response
```

### Correlação de Pearson

A plotagem de correlação se baseia em uma **matriz de correlação** que devemos calcular primeiro.

```python
corr = df[features_response].corr()

corr
```

Observe que essa **correlação** é a **correlação linear**, também conhecida como **correlação de Pearson**.

A correlação de Pearson é usada para medir a força e a direção (isto é, positiva ou negativa) do relacionamento linear entre duas variáveis.

```python
# usando a seaborn para plotar a correlação

sns.heatmap(corr, 
            xticklabels=corr.columns.values, 
            yticklabels=corr.columns.values, 
            center=0)
```

O que essa plotagem mostra?

Em geral, se duas características (ou uma característica e a resposta) estiverem **altamente correlacionadas**, podemos dizer que há uma forte associação entre elas.

As características que estiverem altamente correlacionadas com a resposta serão úteis na previsão.

A plotagem mostra, assim, que a característica **PAY_1** é a mais fortemente correlacinada com a variável de resposta. Também notamos que as características **BILL_AMT** estão altamente correlacionadas entre si.

### Atenção

Embora muito usada, a **correlação de Pearson** não é muito apropriada para uma variável binária, como a que temos em nosso problema.

Tecnicamente falando, entre outras restrições, ela só é válida para **dados contínuos**. 

No entanto, pode ajudar a dar uma visão geral da utilidade das características.

# Teste F

Embora teoricamente a correlação de Pearson seja válida para variáveis de resposta contínuas, a variável de resposta binária dos dados do estudo de caso pode ser considerada um dado categórico com apenas duas categorias: 0 e 1.

Entre todos os testes que podem ser executados para ver se as características estão assocadas a uma resposta categórica, temos o **teste F ANOVA** (abreviação de 'analysis of variance'). O teste F ANOVA pode ser comparado com o **teste F de regressão**, que é muito semelhante à correlação de Pearson.

A saída de um teste F ANOVA é composta da **estatística f** e de **valores-p**. 

O valor-p é um conceito útil para várias medidas estatísticas. Por exemplo, embora não as tenhamos examinado, cada uma das correlações de Pearson calculadas para a matriz de correlação anterior tem um valor-p correspondente.

Há um conceito de valor-p semelhante aplicável aos coeficientes da regressão lienar, aos coeficientes da regressão logística e as outras medidas.
___
No contexto do teste F, o valor-p  responde à seguinte pergunta:

*Qual é a probabilidade de o valor da média dessa característica nas amosras da classe positiva ser IGUAL ao de amostras da classe negativa?*

Se uma característica tiver valores de média muito diferentes entre as classes positivas e negativas:

* é improvável que os valores da média sejam iguais (valor-p baixo);
* provavelmente será uma boa característica em nosso modelo porque nos ajudará a distinguir as classes positivas das negativas
